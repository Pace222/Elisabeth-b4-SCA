# TVLA

```python
traces_path = "..\\acquisition\\125000_same_varying-new_sboxes\\carto_eB4-Rnd-3-WhiteningAndFullFilter-Same-And-Varying-Masking-Shuffling.mat"
key_path = "..\\acquisition\\125000_same_varying-new_sboxes\\carto_eB4-Rnd-3-WhiteningAndFullFilter-Same-And-Varying-Masking-Shuffling.log"
```

```python
TRACE_SIZE = 25_000
data_loader = EntireTraceIterator(traces_path, key_path, nr_populations=2, nr_scenarios=1, trace_size=TRACE_SIZE, traces_per_division=250_000, parse_output="")
for seeds, traces, key, _ in data_loader((0, 1), (0,)):
    traces_same, seeds_same = traces[0][0], seeds[0][0]
    traces_varying, seeds_varying = traces[1][0], seeds[1][0]
```

```python
t_test = ttest_ind(traces_same, traces_varying, axis=0, equal_var=False)
```

```python
np.where(np.abs(t_test.statistic) > 4.5)[0].shape[0]
```

```python
t_test.pvalue.shape[0]
```

```python
fig = px.line(t_test.statistic)
fig.for_each_trace(lambda t: t.update(name="t-test"))
fig.add_hline(4.5, line_color="red")
fig.add_hline(-4.5, line_color="red")

config = {'scrollZoom': True}
#fig.update_layout(dragmode="pan")
fig.update_layout(
    title={
        "text": "TVLA on last round of Elisabeth-b4 with 2-share masking and shuffling enabled",
        "x": 0.5,
        "xanchor": "center",
        "yanchor": "top"
    },
    xaxis={
        "range": [11400, 24810],
        "tickmode": "array",
        "tickvals": list(range(11400, 24820, 671)),
        "ticktext": [str(round(363.6 + 1.34 * n, 1)) for n in range(20)]
    },
    xaxis_title="Time (s)",
    yaxis_range=[-80, 80],
    yaxis_title="t-value",
    showlegend=False
)

fig.show(config=config)
```
