```python
import numpy as np
import pickle as pic
```

```python
from tensorflow.keras.models import load_model
```

```python
with open("splitted_rws.pic", "rb") as r:
    X_training, X_val, X_extraction = pic.load(r)
    X_training, X_val, X_extraction = X_training[:, 150:20000], X_val[:, 150:20000], X_extraction[:, 150:20000]
```

```python
model = load_model("./resnet_models/resnet_750000_orig_rws_perm.keras")
y_prediction = np.zeros((X_extraction.shape[0], 98), dtype=np.float64)
NR = 20
SUB = X_extraction.shape[0] // NR
for i in range(0, X_extraction.shape[0], SUB):
    y_prediction[i:i+SUB] = model.predict(X_extraction[i:i+SUB])
```

```python
y_prediction = np.log(y_prediction)
```

```python
with open("y_extraction_750000_orig_rws_perm.pic", "wb") as w:
    pic.dump(y_prediction, w)
```
