```python
import os
os.environ['CUDA_DEVICE_ORDER'] = "PCI_BUS_ID"
os.environ['CUDA_VISIBLE_DEVICES'] = "6"
```

```python
import numpy as np
import pickle as pic
```

```python
from tensorflow.keras.models import load_model
```

```python
with open("splitted_rws.pic", "rb") as r:
    X_training, X_val, X_extraction = pic.load(r)
    X_training, X_val, X_extraction = X_training[:, 5075:9930], X_val[:, 5075:9930], X_extraction[:, 5075:9930]
```

```python
model = load_model("./resnet_models/resnet_750000_orig_rws_1.keras")
X_extraction = np.pad(X_extraction, ((0, 0), (model.input.shape[1] - X_extraction.shape[1], 0)))
y_prediction = model.predict(X_extraction)
y_prediction = [np.log(pred) for pred in y_prediction]
```

```python
with open("y_extraction_750000_orig_rws_1.pic", "wb") as w:
    pic.dump(y_prediction, w)
```
