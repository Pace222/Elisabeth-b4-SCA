```python
import numpy as np
import pickle as pic
```

```python
from tensorflow.keras.models import load_model
```

```python
ROUND_IDX = 0
assert 0 <= ROUND_IDX < 14
```

```python
with open(f"../traces_and_splits/splitted_round_{ROUND_IDX}.pic", "rb") as r:
    X_training, X_val, X_extraction = pic.load(r)
```

```python
model = load_model(f"resnet_models/resnet_750000_round_{ROUND_IDX}.keras")
y_prediction = model.predict(X_extraction)
y_prediction = [np.log(pred) for pred in y_prediction]
```

```python
with open(f"y_extraction/y_extraction_750000_round_{ROUND_IDX}.pic", "wb") as w:
    pic.dump(y_prediction, w)
```
