```python
import os
os.environ['CUDA_DEVICE_ORDER'] = "PCI_BUS_ID"
os.environ['CUDA_VISIBLE_DEVICES'] = "7"
```

```python
import numpy as np
import pickle as pic
```

```python
from tensorflow.keras.models import load_model
```

```python
with open("../traces_and_splits/splitted_rws.pic", "rb") as r:
    X_training, X_val, X_extraction = pic.load(r)
    X_training, X_val, X_extraction = X_training[:, 150:20000], X_val[:, 150:20000], X_extraction[:, 150:20000]
```

```python
model = load_model("resnet_models/resnet_750000_rws_perm.keras")
y_prediction = model.predict(X_extraction)
y_prediction = [np.log(pred) for pred in y_prediction]
```

```python
with open("y_extraction/y_extraction_750000_rws_perm.pic", "wb") as w:
    pic.dump(y_prediction, w)
```
