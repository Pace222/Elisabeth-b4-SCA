```python
import os
os.environ['CUDA_DEVICE_ORDER'] = "PCI_BUS_ID"
os.environ['CUDA_VISIBLE_DEVICES'] = "7"
```

```python
import numpy as np
import pickle as pic
```

```python
from tensorflow.keras.models import load_model
```

```python
RWS_QUARTER = 0
assert 0 <= RWS_QUARTER < 4
```

```python
WINDOWS = [200, 5075, 9930, 14790, 20060]
with open("../traces_and_splits/splitted_rws.pic", "rb") as r:
    X_training, X_val, X_extraction = pic.load(r)
    X_training, X_val, X_extraction = X_training[:, WINDOWS[RWS_QUARTER]:WINDOWS[RWS_QUARTER+1]], X_val[:, WINDOWS[RWS_QUARTER]:WINDOWS[RWS_QUARTER+1]], X_extraction[:, WINDOWS[RWS_QUARTER]:WINDOWS[RWS_QUARTER+1]]
```

```python
model = load_model(f"resnet_models/resnet_750000{RWS_QUARTER}.keras")
y_prediction = model.predict(X_extraction)
y_prediction = [np.log(pred) for pred in y_prediction]
```

```python
with open(f"y_extraction/y_extraction_750000_rws_{RWS_QUARTER}.pic", "wb") as w:
    pic.dump(y_prediction, w)
```
